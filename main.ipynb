```python
import logging
from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    col, input_file_name, regexp_replace, regexp_extract, split, element_at,
    avg, count, desc, asc, round as spark_round
)
import os
import time
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
from pyspark.sql.functions import countDistinct, collect_set, abs, sum as spark_sum
from pyspark.sql.window import Window

# Configurer le logger
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def init_spark():
    logging.info("Initialisation de la session Spark.")
    return SparkSession.builder \
        .appName("LoL Champions Analysis") \
        .getOrCreate()

def load_files_separately(spark, data_dir, max_files=None):
    try:
        # Construire le chemin complet pour le répertoire de données
        full_path = os.path.abspath(data_dir)
        logging.info(f"Chargement des fichiers CSV depuis : {full_path}")
        
        # Lister les fichiers dans le répertoire
        csv_files = [f for f in os.listdir(full_path) if f.endswith('.csv')]
        if max_files:
            csv_files = csv_files[:max_files]
        
        logging.info(f"{len(csv_files)} fichiers CSV trouvés.")

        # Charger les fichiers un par un
        dataframes = []
        for idx, file_name in enumerate(csv_files):
            file_path = os.path.join(full_path, file_name)
            logging.info(f"Lecture du fichier {idx + 1}/{len(csv_files)} : {file_name}")
            start_time = time.time()
            try:
                # Charger un fichier CSV avec le bon délimiteur
                df = spark.read.csv(file_path, header=True, inferSchema=True, sep=";")
                df = df.withColumn("file_name", input_file_name())

                # Remplacer les encodages %20 par des espaces
                df = df.withColumn("file_name_decoded", regexp_replace(col("file_name"), '%20', ' '))

                # Extraire le nom de fichier de base
                df = df.withColumn("base_file_name", element_at(split(col("file_name_decoded"), "[/\\\\]"), -1))

                # Extraire la version à partir du nom de fichier de base
                df = df.withColumn("version", regexp_extract(col("base_file_name"), r"(\d+\.\d+)\.csv$", 1))

                dataframes.append(df)
                logging.info(f"Le fichier {file_name} a été chargé avec succès.")
            except Exception as e:
                logging.error(f"Erreur lors de la lecture du fichier {file_name}: {e}")
            end_time = time.time()
            
            # Limite de temps pour la lecture de chaque fichier
            elapsed_time = end_time - start_time
            logging.info(f"Temps écoulé pour lire {file_name} : {elapsed_time:.2f} secondes.")
            if elapsed_time > 10:  # Par exemple, arrêter si un fichier prend plus de 10 secondes à charger
                logging.warning(f"Le fichier {file_name} prend trop de temps à charger. Arrêt après {elapsed_time:.2f} secondes.")
                break
        
        # Combiner les DataFrames
        if dataframes:
            logging.info("Combinaison des fichiers CSV dans un seul DataFrame.")
            final_df = dataframes[0]
            for df in dataframes[1:]:
                final_df = final_df.unionByName(df)
            return final_df
        else:
            logging.error("Aucun fichier CSV n'a été chargé.")
            return None

    except Exception as e:
        logging.error(f"Erreur lors du chargement des fichiers : {e}")
        raise e

def perform_eda(df):
    logging.info("Début de l'analyse exploratoire des données...")

    # Nettoyage des colonnes de pourcentage
    logging.info("Nettoyage des colonnes de pourcentage...")

    percentage_columns = ["Win %", "Role %", "Pick %", "Ban %"]

    for col_name in percentage_columns:
        df = df.withColumn(col_name, regexp_replace(col_name, "%", ""))
        df = df.withColumn(col_name, col(col_name).cast("double"))

    # Vérifier les types de données après conversion
    logging.info("Types de données après conversion :")
    df.printSchema()

    # 1. Statistiques globales
    logging.info("Calcul des statistiques globales...")


    # Calcul de la proportion de chaque rang par rôle
    logging.info("Calcul de la proportion de chaque rang par rôle...")

    # Vérifier le type de la colonne 'Tier'
    df = df.withColumn("Tier", col("Tier").cast("string"))

    # Compter le nombre total d'occurrences pour chaque combinaison Role-Tier
    role_tier_counts = df.groupBy("Role", "Tier").agg(count("*").alias("count"))

    # Calculer le total pour chaque rôle
    role_window = Window.partitionBy("Role")
    role_tier_proportions = role_tier_counts.withColumn(
        "total_role_count", spark_sum("count").over(role_window)
    ).withColumn(
        "proportion", spark_round((col("count") / col("total_role_count")) * 100, 2)
    ).orderBy("Role", desc("proportion"))

    # Afficher les résultats
    role_tier_proportions.show(100, truncate=False)

    # Visualisation de la proportion de rangs par rôle
    role_tier_proportions_pd = role_tier_proportions.toPandas()

    # Définir une palette de couleurs manuelle pour les tiers
    tier_colors = {
        'God':'gold',
        'S': 'lightcoral',
        'A': 'lightgreen',
        'B': 'skyblue',
        'C': 'lightgray',
        'D': 'gray'
    }
    
    plt.figure(figsize=(15, 10))
    roles = role_tier_proportions_pd['Role'].unique()
    num_roles = len(roles)
    
    for idx, role in enumerate(roles, 1):
        plt.subplot(3, 2, idx)  # Ajustez ces valeurs selon le nombre de rôles
        role_data = role_tier_proportions_pd[role_tier_proportions_pd['Role'] == role]
        
        # Obtenir les couleurs en fonction des tiers
        colors = [tier_colors.get(tier, 'gray') for tier in role_data['Tier']]

        # Créer le pie chart avec la coloration par tier
        plt.pie(role_data['proportion'], labels=role_data['Tier'], autopct='%1.1f%%', colors=colors)
        plt.title(f"Proportion des rangs pour {role}")
    
    plt.tight_layout()
    plt.savefig("rank_proportion_by_role.png")
    plt.show()
    logging.info("Graphique de la proportion des rangs par rôle sauvegardé sous 'rank_proportion_by_role.png'")


    # Colonnes numériques
    numeric_cols = ["Score", "Trend", "Win %", "Role %", "Pick %", "Ban %", "KDA"]

    # Calculer les statistiques numériques
    stats = df.select([avg(c).alias(f"avg_{c}") for c in numeric_cols])
    stats.show()
    logging.info("Calcul des statistiques globales terminé.")

    # Fréquence des Class, Role et Tier
    logging.info("Calcul de la fréquence des 'Class', 'Role' et 'Tier'...")
    categorical_cols = ["Class", "Role", "Tier"]

    for col_name in categorical_cols:
        logging.info(f"Fréquence de la colonne '{col_name}':")
        freq = df.groupBy(col_name).count().orderBy(desc("count"))
        freq.show(truncate=False)

        # Visualisation des fréquences
        freq_pd = freq.toPandas()
        plt.figure(figsize=(10, 6))
        sns.barplot(x=col_name, y='count', data=freq_pd)
        plt.title(f"Fréquence des '{col_name}'")
        plt.xlabel(col_name)
        plt.ylabel("Nombre d'occurrences")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(f"{col_name}_frequency.png")
        plt.show()
        logging.info(f"Graphique des fréquences pour '{col_name}' sauvegardé sous '{col_name}_frequency.png'")

    # 2. Évolution des statistiques à travers les versions
    logging.info("Analyse de l'évolution des statistiques à travers les versions...")
    version_stats = df.groupBy("version").agg(
        avg("Win %").alias("avg_win_rate"),
        avg("Pick %").alias("avg_pick_rate"),
        avg("Ban %").alias("avg_ban_rate")
    ).orderBy("version")
    version_stats.show()
    logging.info("Analyse de l'évolution des statistiques à travers les versions terminée.")

    # Visualisation de l'évolution des statistiques à travers les versions
    version_stats_pd = version_stats.toPandas()
    plt.figure(figsize=(12, 6))
    sns.lineplot(x='version', y='avg_win_rate', data=version_stats_pd, marker='o', label='Win Rate')
    sns.lineplot(x='version', y='avg_pick_rate', data=version_stats_pd, marker='o', label='Pick Rate')
    sns.lineplot(x='version', y='avg_ban_rate', data=version_stats_pd, marker='o', label='Ban Rate')
    plt.title("Évolution des statistiques moyennes à travers les versions")
    plt.xlabel("Version")
    plt.ylabel("Pourcentage moyen (%)")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.savefig("version_stats_evolution.png")
    plt.show()
    logging.info("Graphique de l'évolution des statistiques à travers les versions sauvegardé sous 'version_stats_evolution.png'")

    # 3. Moyennes des statistiques par Role, Class et Tier
    logging.info("Calcul des moyennes des statistiques par 'Role', 'Class' et 'Tier'...")

    for col_name in categorical_cols:
        logging.info(f"Moyennes des statistiques par '{col_name}':")
        stats_by_cat = df.groupBy(col_name).agg(
            avg("Win %").alias("avg_win_rate"),
            avg("Pick %").alias("avg_pick_rate"),
            avg("Ban %").alias("avg_ban_rate"),
            avg("KDA").alias("avg_kda")
        ).orderBy(desc("avg_win_rate"))
        stats_by_cat.show(truncate=False)

        # Visualisation des moyennes par catégorie
        stats_by_cat_pd = stats_by_cat.toPandas()
        plt.figure(figsize=(10, 6))
        sns.barplot(x=col_name, y='avg_win_rate', data=stats_by_cat_pd)
        plt.title(f"Taux de victoire moyen par '{col_name}'")
        plt.xlabel(col_name)
        plt.ylabel("Taux de victoire moyen (%)")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(f"{col_name}_win_rate.png")
        plt.show()
        logging.info(f"Graphique du taux de victoire moyen par '{col_name}' sauvegardé sous '{col_name}_win_rate.png'")

    # 4. Top 10 des champions avec le meilleur taux de victoire
    logging.info("Identification des 10 meilleurs champions par taux de victoire...")
    top_win_rate = df.orderBy(desc("Win %")).select("Name", "Win %", "version").limit(10)
    top_win_rate.show()
    logging.info("Identification des 10 meilleurs champions par taux de victoire terminée.")

    # 5. Top 10 des champions avec le pire taux de victoire
    logging.info("Identification des 10 pires champions par taux de victoire...")
    worst_win_rate = df.orderBy(asc("Win %")).select("Name", "Win %", "version").limit(10)
    worst_win_rate.show()
    logging.info("Identification des 10 pires champions par taux de victoire terminée.")

    # 6. Analyse de la corrélation entre les statistiques
    logging.info("Calcul de la matrice de corrélation...")
    correlation_matrix = df.select(numeric_cols).toPandas().corr()

    # Visualisation de la matrice de corrélation
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
    plt.title("Matrice de corrélation des statistiques des champions")
    plt.tight_layout()
    plt.savefig("correlation_matrix.png")
    plt.show()
    logging.info("Matrice de corrélation sauvegardée sous 'correlation_matrix.png'")

    # 7. Analyse de la distribution des taux de victoire
    logging.info("Analyse de la distribution des taux de victoire...")
    win_rates = df.select("Win %").toPandas()
    plt.figure(figsize=(10, 6))
    sns.histplot(win_rates["Win %"], kde=True)
    plt.title("Distribution des taux de victoire des champions")
    plt.xlabel("Taux de victoire (%)")
    plt.ylabel("Fréquence")
    plt.tight_layout()
    plt.savefig("win_rate_distribution.png")
    plt.show()
    logging.info("Distribution des taux de victoire sauvegardée sous 'win_rate_distribution.png'")

    # Win % en fonction des classes
    logging.info("Calcul du taux de victoire moyen par classe...")
    win_by_class = df.groupBy("Class").agg(
        avg("Win %").alias("avg_win_rate")
    ).orderBy(desc("avg_win_rate"))
    win_by_class.show()

    # Visualisation du taux de victoire par classe
    win_by_class_pd = win_by_class.toPandas()
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Class', y='avg_win_rate', data=win_by_class_pd)
    plt.title("Taux de victoire moyen par classe de champion")
    plt.xlabel("Classe")
    plt.ylabel("Taux de victoire moyen (%)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig("win_rate_by_class.png")
    plt.show()

    # Calcul de la proportion de chaque rang par rôle
    logging.info("Calcul de la proportion de chaque rang par rôle...")
    
    # Compter le nombre total d'occurrences pour chaque combinaison Role-Tier
    role_tier_counts = df.groupBy("Role", "Tier").agg(count("*").alias("count"))
    
    # Calculer le total pour chaque rôle
    role_window = Window.partitionBy("Role")
    role_tier_proportions = role_tier_counts.withColumn(
        "total_role_count", sum("count").over(role_window)
    ).withColumn(
        "proportion", spark_round((col("count") / col("total_role_count")) * 100, 2)
    ).orderBy("Role", desc("proportion"))

    # Afficher les résultats
    role_tier_proportions.show(100, truncate=False)

    # Visualisation de la proportion de rangs par rôle
    role_tier_proportions_pd = role_tier_proportions.toPandas()
    
    plt.figure(figsize=(15, 10))
    roles = role_tier_proportions_pd['Role'].unique()
    num_roles = len(roles)
    
    for idx, role in enumerate(roles, 1):
        plt.subplot(3, 2, idx)  # Ajustez ces valeurs selon le nombre de rôles
        role_data = role_tier_proportions_pd[role_tier_proportions_pd['Role'] == role]
        plt.pie(role_data['proportion'], labels=role_data['Tier'], autopct='%1.1f%%')
        plt.title(f"Proportion des rangs pour {role}")
    
    plt.tight_layout()
    plt.savefig("rank_proportion_by_role.png")
    plt.show()
    logging.info("Graphique de la proportion des rangs par rôle sauvegardé sous 'rank_proportion_by_role.png'")


    # 1. Champions les plus polyvalents (ayant le plus de rôles)
    logging.info("Identification des champions les plus polyvalents...")
    polyvalent_champions = df.groupBy("Name").agg(
        countDistinct("Role").alias("role_count"),
        collect_set("Role").alias("roles")
    ).orderBy(desc("role_count"))
    polyvalent_champions.show(10, truncate=False)

    # 2. Évolution du taux de victoire des champions à travers les versions
    logging.info("Analyse de l'évolution du taux de victoire des champions...")
    win_rate_evolution = df.groupBy("Name", "version").agg(
        avg("Win %").alias("avg_win_rate")
    ).orderBy("Name", "version")
    win_rate_evolution.show()

    # Visualisation de l'évolution du taux de victoire pour les 5 premiers champions
    top_5_champions = win_rate_evolution.select("Name").distinct().limit(5).collect()
    top_5_names = [row["Name"] for row in top_5_champions]
    
    win_rate_evolution_top5 = win_rate_evolution.filter(col("Name").isin(top_5_names))
    win_rate_evolution_top5_pd = win_rate_evolution_top5.toPandas()
    
    plt.figure(figsize=(12, 6))
    for name in top_5_names:
        data = win_rate_evolution_top5_pd[win_rate_evolution_top5_pd["Name"] == name]
        plt.plot(data["version"], data["avg_win_rate"], marker='o', label=name)
    
    plt.title("Évolution du taux de victoire pour 5 champions")
    plt.xlabel("Version")
    plt.ylabel("Taux de victoire moyen (%)")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig("win_rate_evolution_top5.png")
    plt.show()

    # 3. Champions avec le plus grand écart entre leur taux de sélection et leur taux de bannissement
    logging.info("Identification des champions avec le plus grand écart entre sélection et bannissement...")
    pick_ban_difference = df.withColumn(
        "pick_ban_diff",
        abs(col("Pick %") - col("Ban %"))
    ).orderBy(desc("pick_ban_diff"))
    pick_ban_difference.select("Name", "Pick %", "Ban %", "pick_ban_diff").show(10)

    # 4. Analyse de la relation entre le KDA et le taux de victoire
    logging.info("Analyse de la relation entre KDA et taux de victoire...")
    kda_win_relation = df.select("Name", "KDA", "Win %")
    kda_win_relation_pd = kda_win_relation.toPandas()
    
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x="KDA", y="Win %", data=kda_win_relation_pd)
    plt.title("Relation entre KDA et taux de victoire")
    plt.xlabel("KDA")
    plt.ylabel("Taux de victoire (%)")
    plt.tight_layout()
    plt.savefig("kda_win_rate_relation.png")
    plt.show()

    logging.info("Analyse exploratoire des données terminée.")

def main():
    # Initialiser Spark
    spark = init_spark()
    try:
        data_dir = r"C:\Users\brice\Desktop\M2\M2_S1\Donnees_Massives\Projet_SPARK\League_of_PySpark\data"
        
        # Charger les fichiers un par un
        df = load_files_separately(spark, data_dir, max_files=10)  # Par exemple, ne charger que 10 fichiers
        
        if df:
            # Afficher le schéma et quelques lignes pour vérifier
            df.printSchema()
            df.show(5)

            # Exemple d'opération simple
            champion_count = df.select("Name").distinct().count()
            logging.info(f"Nombre total de champions uniques : {champion_count}")
            
            # Afficher le nombre total de lignes
            total_rows = df.count()
            logging.info(f"Nombre total de lignes : {total_rows}")

            # Appel à la fonction d'analyse exploratoire des données
            perform_eda(df)
    except Exception as e:
        logging.error(f"Erreur rencontrée pendant l'exécution : {e}")
    finally:
        logging.info("Arrêt de la session Spark...")
        spark.stop()

if __name__ == "__main__":
    main()
```
